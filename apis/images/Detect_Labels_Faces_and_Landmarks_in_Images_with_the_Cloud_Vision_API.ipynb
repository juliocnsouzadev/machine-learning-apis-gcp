### Label Detection

```
import requests
pkey = {'key': API_KEY}
img = 'https://www.biggerbolderbaking.com/wp-content/uploads/2018/05/NO-yeast-Homemade-donuts-FINAL.jpg'
```

```
def callApi(feature, img):
  url = 'https://vision.googleapis.com/v1/images:annotate?key=' + API_KEY
  requestObject={'requests':[{'image':{'source':{'imageUri':img}},'features':[{'type':feature,'maxResults':10}]}]}
  x = requests.post(url,params = pkey, json = requestObject)
  print(x.text)
```

```
callApi('LABEL_DETECTION',img)
```

### Web Detection

```
callApi("WEB_DETECTION", img)
```

### Face Detection
